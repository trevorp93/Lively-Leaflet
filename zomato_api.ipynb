```python
#initialize arrays to hold restaurant info 
restaurant_ids = []
restaurant_name=[]
restaurant_address=[]
restaurant_latitude=[]
restaurant_longitude=[]
restaurant_cuisine=[]
restaurant_price_range=[]
restaurant_aggregate_rating=[]

#initialize arrays to hold review info 
restaurant_review_text=[]
restaurant_review_rating=[]
restaurant_review_id=[]
restaurant_review_res_id=[]
```

```python
#main code to retrieve zomato restaurants and reviews and append to lists created above

#cuisine ids for american, indian, chinese, mexican and italian
#cuisines=["1","148","25","73","55"]
cuisines=["25","73","55"]
#url to pull restaurants for a specific cuisine type
restuarant_url="https://developers.zomato.com/api/v2.1/search?entity_id=280&entity_type=city&cuisines="
#url to pull reviews for a specific restaurant id
review_url = "https://developers.zomato.com/api/v2.1/reviews?res_id="
api_key = "&apikey=9be961b0229785ac06ff1c9d786e1560"
#api_key = "&apikey=bf0e68dc73ded0601b690c7969d9ae40"
count1="&count=100"
count2="&count=10"

#loop to get restuarants for NYC restaurants for 5 types for cuisines
for cuisine in cuisines:

    restaurants = req.get(restuarant_url+cuisine+api_key+count1).json()
     #loop to get restaurant info per restaurant and save into arrays
    try:
        for restaurant in restaurants["restaurants"]:
            print("%s | %s | %s | %s |%s|%s|%s|%s" % (restaurant["restaurant"]["name"], restaurant["restaurant"]["cuisines"],restaurant["restaurant"]["R"]["res_id"], restaurant["restaurant"]["location"]["address"],restaurant["restaurant"]["location"]["latitude"],restaurant["restaurant"]["location"]["longitude"], restaurant["restaurant"]["price_range"], restaurant["restaurant"]["user_rating"]["aggregate_rating"] ))                                                  
            restaurant_ids.append(int(restaurant["restaurant"]["R"]["res_id"]))
            restaurant_name.append(restaurant["restaurant"]["name"])
            restaurant_address.append(restaurant["restaurant"]["location"]["address"])
            restaurant_latitude.append(restaurant["restaurant"]["location"]["latitude"])
            restaurant_longitude.append(restaurant["restaurant"]["location"]["longitude"])
            restaurant_cuisine.append(restaurant["restaurant"]["cuisines"])
            restaurant_price_range.append(restaurant["restaurant"]["price_range"])
            restaurant_aggregate_rating.append(restaurant["restaurant"]["user_rating"]["aggregate_rating"])  
        
            #loop to get reviews per restaurant id
            for res_id in restaurant_ids:
                res_reviews = req.get(review_url + str(res_id) + api_key + count2).json()
                #print(json.dumps(res_reviews))
        
                #loop to save reviews into arrays
            
                try:
                    for review in res_reviews["user_reviews"]:
                        print("%s | %s | %s " % (review["review"]["rating"], review["review"]["review_text"],review["review"]["id"] ))                                                                        
                        restaurant_review_text.append(review["review"]["review_text"])
                        restaurant_review_rating.append(review["review"]["rating"])
                        restaurant_review_id.append(review["review"]["id"])
                        restaurant_review_res_id.append(int(res_id))
                except KeyError:
                    print("Oops, review doesn't exist. Hey, but the code didn't die, moving forward")
    except KeyError:
        print("Oops, restaurant doesn't exist. Hey, but the code didn't die, moving forward")
                
```

```python
#create pd for restaurants
restaurant_list = pd.DataFrame(
{
'restaurant_id' : restaurant_ids,
'restaurant_name' : restaurant_name,
'restaurant_cuisine' : restaurant_cuisine,
'restaurant_price_range' : restaurant_price_range,
'restaurant_aggregate_rating' : restaurant_aggregate_rating,
'restaurant_address' : restaurant_address,
'restaurant_latitude' : restaurant_latitude,
'restaurant_longitude' : restaurant_longitude
}
)

restaurant_list
```

```python
#create pd for reviews
reviews_list = pd.DataFrame(
{
'restaurant_id' : restaurant_review_res_id,
'restaurant_review_text' :  restaurant_review_text,
'restaurant_review_rating' : restaurant_review_rating,
'restaurant_review_id' : restaurant_review_id
}
)

reviews_list
```

```python
# create combined pd with restuarant info and reviews
result = pd.merge(restaurant_list, reviews_list, how='left' , on='restaurant_id')
result
```
